```python
import pandas as pd
```

# Finding the nulls

```python
df=pd.read_csv('dataset.csv')
df.isna().sum()
```

# Filling the nulls in seniorcitizen with mean

```python
mean=df.SeniorCitizen.mean()
df.SeniorCitizen.fillna(mean, inplace=True)
```

# Filling the nulls in tenure with zeros

```python
df.tenure.fillna(0,inplace=True)
```

# Making sure that nulls no longer exist

```python
df.isna().sum()
```

# Dealing with Outliers in Tenure

```python
df.head()
```

```python
df['tenure']=round(df.TotalCharges/df.MonthlyCharges,0).astype(int)
df.head()
```

# Label Encoding (Categorical Data)

```python
newDf=df[['gender','Partner','PhoneService','MultipleLines','InternetService']]
newDf.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_gender=LabelEncoder()
le_part=LabelEncoder()
le_serv=LabelEncoder()
le_lines=LabelEncoder()
le_internet=LabelEncoder()
```

```python
newDf['gender']=le_gender.fit_transform(newDf.gender)
newDf['Partner']=le_part.fit_transform(newDf.Partner)
newDf['PhoneService']=le_serv.fit_transform(newDf.PhoneService)
newDf['MultipleLines']=le_lines.fit_transform(newDf.MultipleLines)
newDf['InternetService']=le_internet.fit_transform(newDf.InternetService)
```

```python
newDf.head()
```

# Feature Scaling

```python
df['tenure']=(df.tenure-df.tenure.min())/(df.tenure.max()-df.tenure.min())
```

```python
df['MonthlyCharges']=(df.MonthlyCharges-df.MonthlyCharges.min())/(df.MonthlyCharges.max()-df.MonthlyCharges.min())
```

```python
df['TotalCharges']=(df.TotalCharges-df.TotalCharges.min())/(df.TotalCharges.max()-df.TotalCharges.min())
```

```python
df.MonthlyCharges.max(),df.MonthlyCharges.min()
```

# Feature Selection

```python
#Here we don't need customer ID , StreamingMovies so we'll drop these 2 columns
df.drop('customerID',axis=1, inplace=True)
df.drop('StreamingMovies',axis=1, inplace=True)
```
